```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', size=12)

data_connect = np.loadtxt('connect_H0.txt', usecols=[0, 1], skiprows=1)
data_prospect = np.loadtxt('prospect_H0.txt', usecols=[0, 2], skiprows=1)
data_class = np.loadtxt('class_H0.txt', usecols=[0, 2], skiprows=1)

fig, ax = plt.subplots()
ax.tick_params(direction='in', left=False, labelleft=False)
ax.set_xlim(66.5, 70)
ax.set_xlabel(r'$\mathbf{H}_{0}$')
ax.set_ylabel(r'$\mathcal{L}(\mathbf{H}_{0})/\mathcal{L}_{max}$')
ax.plot(data_connect[:, 0], np.exp(-data_connect[:, 1]) / max(np.exp(-data_connect[:, 1])), marker='*', label='Connect')
ax.plot(data_prospect[:, 0], np.exp(-data_prospect[:, 1]) / max(np.exp(-data_prospect[:, 1])), marker='.', label='Prospect(Connect)')
ax.plot(data_class[:, 0], np.exp(-data_class[:, 1]) / max(np.exp(-data_class[:, 1])), marker='o', label='Prospect(Class)')
ax.legend(prop={'family': 'monospace'})  # Set font family for legend

plt.savefig('prospect_comparison.pdf')
```
